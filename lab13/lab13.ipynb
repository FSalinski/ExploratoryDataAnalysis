## Figury

`plt.plot()` pozwala na narysowanie tylko jednej figury o współrzędnych (x,y)

```python
names = ['X', 'Y', 'Z']
values = [3, 9, 27]
```

```python
plt.bar(names, values)
plt.show()
```

`plt.figure()` definiuje figurę, możemy na niej umieszczać inne elementy

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1) #nrow, ncol, index
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1) #nrow, ncol, index
ax.bar(names, values)
plt.show()
```

`plt.subplot()` - pozwala na rysowanie jednej lub kilku figur w tym samym oknie

```python
fig, ax = plt.subplots(2, 2) #2 wiersze, 1 kolumna
```

```python
fig, ax = plt.subplots(2, 1)
fig.suptitle('Tytuł')

ax[0].bar(names, values)
ax[1].scatter(names, values)
plt.show()
```

# seaborn

```python
import seaborn as sns
import pandas as pd
```

```python
# opcje wyświetlania
sns.set(rc = {'figure.figsize': (12,7)})
sns.set_style("darkgrid")
pd.options.display.max_rows = 10
pd.set_option('display.width', 5000)
pd.set_option('display.max_columns', 60)
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/mini-pw/2021Z-DataVisualizationTechniques/master/labs/data/Pokemon.csv", index_col=0, sep=",")
```

```python
df.head()
```

## Wykres punktowy

```python
sns.scatterplot(data=df, x='Attack', y='Defense')
plt.show()
```

```python
sns.scatterplot(data=df, x='Attack', y='Defense', hue='Stage')
plt.show()
```

```python
sns.scatterplot(data=df, x='Attack', y='Defense', size='Stage')
plt.show()
```

```python
sns.scatterplot(data=df, x='Attack', y='Defense', hue='Stage')

plt.ylim(None, 150)
plt.xlim(-100, None)
plt.show()
```

## boxplot

```python
sns.boxplot(data=df, color='w')
plt.show()
```

```python
sns.boxplot(data=df, color='w')
plt.show()
```

```python
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)
stats_df.head()
```

```python
sns.boxplot(data=stats_df, color = 'lightblue')
```

```python
sns.boxplot(data=stats_df, x = 'Type 1', y = 'Attack')
plt.show()
```

## violin plot

```python
sns.violinplot(x = 'Type 1', y = 'Attack', data=df)
```

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]

sns.violinplot(x='Type 1', y='Attack', data=df, 
               palette=pkmn_type_colors)
plt.show()

```

## swarm plot

```python
sns.swarmplot(x='Type 1', y='Attack', data=df, 
              palette=pkmn_type_colors)
plt.show()
```

## nakładanie wykresów

```python
sns.violinplot(x='Type 1',
               y='Attack', 
               data=df, 
               inner=None, # usunięcie boxplotów ze środka
               palette=pkmn_type_colors)
 
sns.swarmplot(x='Type 1', 
              y='Attack', 
              data=df, 
              color='black')# czarne punkty
 
# Tytuł z matplotlib
plt.title('Attack by Type')
plt.show()
```

```python
stats_df.shape
```

```python
melted_df = pd.melt(stats_df, 
                    id_vars=["Name", "Type 1", "Type 2"], # Zmienne do zatrzymania
                    var_name="Stat") # Nazwa nowej zmiennej
melted_df
```

```python
sns.boxplot(x='Stat', y='value', data=melted_df, 
              hue='Type 1')
plt.show()
```

```python
# powiększamy wykres
plt.figure(figsize=(16,7))
 
sns.boxplot(x='Stat', 
              y='value', 
              data=melted_df, 
              hue='Type 1', 
              palette=pkmn_type_colors) # dodajemy paletę

# przenosimy legendę
plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```

## panele

Narysować wykresy dla zbioru danych Titanic:

a) histogram zmiennej fare 

b) boxplot wieku w podziale na płeć

c) wykres słupkowy pokazujący liczbę pasażerów w każdej z trzech klas

d) procent osób, które przeżyły (survived) w każdej z klas

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
plt.figure(figsize=(6, 5))
sns.histplot(titanic['fare'], bins=25, color='lightblue')
plt.title('Histogram of Fare', size=16)
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.boxplot(x='sex', y='age', data=titanic)
plt.title('Boxplot of age by gender', size=16)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='class', data=titanic)
plt.title('Number of passengers by class', size=16)
plt.show()
```

```python
df = titanic.groupby('class').agg({'embarked' : 'count', 'survived' : 'sum'})
df['ratio'] = df['survived'] / df['embarked']
df.drop(['embarked', 'survived'], axis=1, inplace=True)
df
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=df.index, y='ratio', data=df,
            color='skyblue')
plt.title('Survival ratio by class', size=16)
plt.ylabel('Survival ratio')
plt.xlabel('Class')
plt.show()
```

e) Narysuj violin plot wieku w zależności od klasy i płci

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='class', hue='sex', split=True, y='age', data=titanic,
               inner='quart')
plt.title('Age distribution by class and sex', size=16)
plt.xlabel('Class')
plt.ylabel('Age')
plt.legend(title='Gender')
plt.show()
```
